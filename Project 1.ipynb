6.) Describe conceptually the function and fundamental principles of the klystron. What are
the key input/output variables and their dependencies?

<ul>
<li> A klystron fundamentally functions as a microwave amplifier driven by a low power microwave source. In the klystron electrons produced by the cathode are accelerated into the first cavity of the klystron energized by the lower power microwaves from original source. This creates an oscillating electric field in the cavity which disproportionately slows down and speeds up the electrons resulting in the bunching. As the bunches of electrons arrive at the second cavity they decelerate and release energy which is converrted to high power microwaves. A diagram of this process is shown in the figure below found in chapter 4 of Khan. 
</li>
<li> The main input variables into the klystron are the klystron voltage and the RF input power, where the main output variable is RF output power.</li>
<li>See: Khan, F. M., & Gibbons, J. P. (2016). Khans the physics of radiation therapy. Philadelphia: Wolters Kluwer.</li>
</ul>

4.) Describe conceptually the function and fundamental principles of the accelerating
waveguide. What are the key input/output variables and their dependencies?

<ul>
<li> Accelerating waveguides consist of metal vaccuum cylinders that are separated into sections by dividers with small holes in the center that allow the electrons to pass through them. An oscillating field is then induced to accelerate the electrons through each of the sections of the waveguide to nearly the speed of light. Waveguides are either standing or traveling based on how the oscillating electric field changes in time.
</li>

<li> The main input variables of an accelerating waveguide are the RF Output from the klystron and the Gun current magnet current. These will determine how the electrons are accelerated and with what energy the beam will have as well as the target current (outputs).
</li>

<li>See: https://oncologymedicalphysics.com/accelerating-waveguides/
</li>
</ul>

5.) Explain some of the differences between a travelling and a standing waveguide.

<ul>
<li> In a traveling waveguide, the electrons travel with the oscillating electric field down the waveguide (amplitude changes with position, with no traveling nodes). However not all of the energy from the electric field is deposited to the electrons so there must be an absorbing resistive material at the ened of the last cavity to prevent reflection of this traveling wave. Below is an animation of a traveling waveguide. </li>

</ul>
![SegmentLocal](TravelingWave-Waveguide1.gif "segment")

<ul>
<li> In a standing waveguide the oscillating electric field changes amplitude with time instead of position. Standing waveguides allow reflection of the electric field at the end of the waveguide and use this backward traveling wave to create a standing waveform (fixed nodes). Here, every other cavity can be used to accelerate electrons instead of four needed by the traveling waveguide. Below is an animation of a standing waveguide.</li>
</ul>
![SegmentLocal](StandingWave_Waveguide.gif "segment")

<ul>
<li> Animations are from: https://oncologymedicalphysics.com/accelerating-waveguides/</li>

b. Build-up is not modeled in the SIMAC software. How does build-up affect PDD
curves in actual beam data? Why does build-up occur?

<ul>
<li> Build up affects actual beam data PDD curves because the highest dose is not actually at the shallowest depth. This means that the PDD is not monotonically decreasing like it seems in the graph above. This makes it so the PDD curve increases at shallow depths, hits a peak dose at a specific depth, and then decreases from there.</li>

<li> Build up occurs because dose comes from charged particles leaving their energy in tissue through interaction. When the photon beam hits tissue and charged particles are produced they deposit their energy as they travel further through the tissue. This means that the dose is actually slightly deeper in the tissue from where the original photon interacted. </li>
</ul>
